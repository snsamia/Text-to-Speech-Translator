```python
from langdetect import detect
from gtts import gTTS
from IPython.display import Audio
from googletrans import Translator

class TextToSpeechTranslator:
    def __init__(self, file_path):
        self.file_path = file_path
        self.text = ""
        self.detected_lang = ""
        self.translated_text = ""
    
    def read_text_file(self):
        with open(self.file_path, "r", encoding="utf-8") as f:
            self.text = f.read()
        print("ğŸ“„ Text loaded successfully.")
    
    def detect_language(self):
        self.detected_lang = detect(self.text)
        print(f"ğŸŒ Detected Language: {self.detected_lang}")
    
    def convert_to_speech(self, output_file="output.mp3"):
        if not self.text or not self.detected_lang:
            raise ValueError("Missing text or language. Run read and detect first.")
        tts = gTTS(text=self.text, lang=self.detected_lang)
        tts.save(output_file)
        print(f"ğŸ”Š Native Speech saved to {output_file}")
        return Audio(output_file)
    
    def translate_to_german(self):
        translator = Translator()
        self.translated_text = translator.translate(self.text, src=self.detected_lang, dest='de').text
        print("ğŸˆ¯ Translated to German (first 300 chars):\n", self.translated_text[:500])
    
    def german_speech(self, output_file="translated_output.mp3"):
        if not self.translated_text:
            raise ValueError("Translated text is empty. Run translate_to_german first.")
        tts = gTTS(text=self.translated_text, lang='de')
        tts.save(output_file)
        print(f"ğŸ”‰ German Speech saved to {output_file}")
        return Audio(output_file)
```

```python
tts = TextToSpeechTranslator("/kaggle/input/bangla-text/bangla.txt")
tts.read_text_file()
tts.detect_language()
tts.convert_to_speech("output_bn.mp3")        

```

```python
tts.translate_to_german()
tts.german_speech("output_de.mp3")            
```
